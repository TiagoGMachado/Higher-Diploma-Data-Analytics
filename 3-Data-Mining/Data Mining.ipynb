Pandas-put data in data frame and manipulate it

numpy-for numerical analysis 

matplotlib-for visualisation 

seaborn-for visualisation(built ontop of matplotlib)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
icu = pd.read_csv("ICU Dataset.csv")
icu
```

```python
icu.dropna() # drop missing values
```

```python
icu.describe() # gives quantitative summary of the data
```

```python
icu.head() # gives top 5 rows
```

```python
icu.tail() # gives last 5 rows
```

```python
variableOfInterest = icu[["age", "bmi", "hr_1st", "map_1st"]]
variableOfInterest
```

```python
correlation = variableOfInterest.corr()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(correlation, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation of ICU Data")
plt.show()
```
